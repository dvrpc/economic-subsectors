```python
import pandas as pd
import numpy as np
```

# Import 2017 Data

```python
raw_2017 = pd.read_csv(r"G:\My Drive\BCarney\DataAnalysis\CBP_2017\NAICS_AllDigits\County_level\rawdata\cbp17co.txt")
raw_2017
```

# Select Three-Digit NAICS

```python
threedigit_NAICS = raw_2017[raw_2017['naics'].str.contains('///')]
threedigit_NAICS
```

# Create new three-digit NAICS column

```python
threedigit_NAICS['naics_threedigit'] = threedigit_NAICS['naics'].str.slice(start=0, stop=3)
threedigit_NAICS
```

# Check for empflag w/ three-digit NAICS

```python
threedigit_NAICS_empflagcheck = threedigit_NAICS[threedigit_NAICS.empflag.notnull()]
threedigit_NAICS_empflagcheck
```

# Select Delaware County

```python
Delaware_NAICS = threedigit_NAICS.loc[(threedigit_NAICS['fipstate'] == 42) & (threedigit_NAICS['fipscty'] == 45)]
Delaware_NAICS
```

```python
Delaware_ThreeDigitNAICS = Delaware_NAICS[['naics_threedigit', 'empflag', 'emp']]
Delaware_ThreeDigitNAICS
```

# Check for empflag

```python
empflag_threedigitcheck = Delaware_ThreeDigitNAICS.loc[Delaware_ThreeDigitNAICS['empflag'].notnull()]
empflag_threedigitcheck
```

# Create midpoint values for empflag rows

```python
Delaware_midpoint = Delaware_ThreeDigitNAICS.assign(emp_fix = Delaware_ThreeDigitNAICS['emp'])
Delaware_midpoint
```

```python
Delaware_midpoint['emp_fix'].loc[(Delaware_midpoint['empflag'] == 'A')] = 10
Delaware_midpoint['emp_fix'].loc[(Delaware_midpoint['empflag'] == 'B')] = 60
Delaware_midpoint['emp_fix'].loc[(Delaware_midpoint['empflag'] == 'C')] = 175
Delaware_midpoint['emp_fix'].loc[(Delaware_midpoint['empflag'] == 'E')] = 375
Delaware_midpoint['emp_fix'].loc[(Delaware_midpoint['empflag'] == 'F')] = 750
Delaware_midpoint['emp_fix'].loc[(Delaware_midpoint['empflag'] == 'G')] = 1750
Delaware_midpoint['emp_fix'].loc[(Delaware_midpoint['empflag'] == 'H')] = 3750
Delaware_midpoint['emp_fix'].loc[(Delaware_midpoint['empflag'] == 'I')] = 7500
Delaware_midpoint['emp_fix'].loc[(Delaware_midpoint['empflag'] == 'J')] = 17500
Delaware_midpoint['emp_fix'].loc[(Delaware_midpoint['empflag'] == 'K')] = 37500
Delaware_midpoint['emp_fix'].loc[(Delaware_midpoint['empflag'] == 'L')] = 75000
Delaware_midpoint['emp_fix'].loc[(Delaware_midpoint['empflag'] == 'M')] = 100000
Delaware_midpoint
```

# View rows where emp_fix does not equal emp

```python
check = Delaware_midpoint.loc[Delaware_midpoint['emp'] != Delaware_midpoint['emp_fix']]
check
```

# Group by 3-digit NAICS, sum employment

```python
Delaware_ThreeDigitNAICS_Totals = Delaware_midpoint.groupby('naics_threedigit').sum()
Delaware_ThreeDigitNAICS_Totals
```

```python
Delaware_ThreeDigitNAICS_Totals.rename(columns = {"emp": "emp_17", "emp_fix": "emp_fix_17"}, inplace=True)
Delaware_ThreeDigitNAICS_Totals
```

# Convert index dtype to str

```python
Delaware_ThreeDigitNAICS_Totals.index = Delaware_ThreeDigitNAICS_Totals.index.astype(str)
Delaware_ThreeDigitNAICS_Totals.index.dtype
```

# Import NAICS index

```python
naics_index = pd.read_csv(r"G:\Shared drives\Community & Economic Development\Economic Clusters\CBP_ThreeDigitNAICS_Index.csv", index_col='NAICS')
naics_index.index = naics_index.index.astype(str)
naics_index.index.dtype
```

# Join two dfs

```python
Delaware_NAICStotals= pd.merge(naics_index, Delaware_ThreeDigitNAICS_Totals, how="left", left_index=True, right_index=True)
Delaware_NAICStotals
```

```python
Delaware_NAICStotals.to_csv(r"G:\My Drive\BCarney\DataAnalysis\CBP_2017\ThreeDigit_NAICS\DVRPC_IndividualCounties\CBP2017_DVRPC_IndividualCounties_Delaware_NAICStotals.csv")
```
