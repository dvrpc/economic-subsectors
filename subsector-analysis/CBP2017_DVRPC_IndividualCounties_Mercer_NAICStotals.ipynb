```python
import pandas as pd
import numpy as np
```

# Import 2017 Data

```python
raw_2017 = pd.read_csv(r"G:\My Drive\BCarney\DataAnalysis\CBP_2017\NAICS_AllDigits\County_level\rawdata\cbp17co.txt")
raw_2017
```

# Select Three-Digit NAICS

```python
threedigit_NAICS = raw_2017[raw_2017['naics'].str.contains('///')]
threedigit_NAICS
```

# Create new three-digit NAICS column

```python
threedigit_NAICS['naics_threedigit'] = threedigit_NAICS['naics'].str.slice(start=0, stop=3)
threedigit_NAICS
```

# Check for empflag w/ three-digit NAICS

```python
threedigit_NAICS_empflagcheck = threedigit_NAICS[threedigit_NAICS.empflag.notnull()]
threedigit_NAICS_empflagcheck
```

# Select Mercer County

```python
Mercer_NAICS = threedigit_NAICS.loc[(threedigit_NAICS['fipstate'] == 34) & (threedigit_NAICS['fipscty'] == 21)]
Mercer_NAICS
```

```python
Mercer_ThreeDigitNAICS = Mercer_NAICS[['naics_threedigit', 'empflag', 'emp']]
Mercer_ThreeDigitNAICS
```

# Check for empflag

```python
empflag_threedigitcheck = Mercer_ThreeDigitNAICS.loc[Mercer_ThreeDigitNAICS['empflag'].notnull()]
empflag_threedigitcheck
```

# Create midpoint values for empflag rows

```python
Mercer_midpoint = Mercer_ThreeDigitNAICS.assign(emp_fix = Mercer_ThreeDigitNAICS['emp'])
Mercer_midpoint
```

```python
Mercer_midpoint['emp_fix'].loc[(Mercer_midpoint['empflag'] == 'A')] = 10
Mercer_midpoint['emp_fix'].loc[(Mercer_midpoint['empflag'] == 'B')] = 60
Mercer_midpoint['emp_fix'].loc[(Mercer_midpoint['empflag'] == 'C')] = 175
Mercer_midpoint['emp_fix'].loc[(Mercer_midpoint['empflag'] == 'E')] = 375
Mercer_midpoint['emp_fix'].loc[(Mercer_midpoint['empflag'] == 'F')] = 750
Mercer_midpoint['emp_fix'].loc[(Mercer_midpoint['empflag'] == 'G')] = 1750
Mercer_midpoint['emp_fix'].loc[(Mercer_midpoint['empflag'] == 'H')] = 3750
Mercer_midpoint['emp_fix'].loc[(Mercer_midpoint['empflag'] == 'I')] = 7500
Mercer_midpoint['emp_fix'].loc[(Mercer_midpoint['empflag'] == 'J')] = 17500
Mercer_midpoint['emp_fix'].loc[(Mercer_midpoint['empflag'] == 'K')] = 37500
Mercer_midpoint['emp_fix'].loc[(Mercer_midpoint['empflag'] == 'L')] = 75000
Mercer_midpoint['emp_fix'].loc[(Mercer_midpoint['empflag'] == 'M')] = 100000
Mercer_midpoint
```

# View rows where emp_fix does not equal emp

```python
check = Mercer_midpoint.loc[Mercer_midpoint['emp'] != Mercer_midpoint['emp_fix']]
check
```

# Group by 3-digit NAICS, sum employment

```python
Mercer_ThreeDigitNAICS_Totals = Mercer_midpoint.groupby('naics_threedigit').sum()
Mercer_ThreeDigitNAICS_Totals
```

```python
Mercer_ThreeDigitNAICS_Totals.rename(columns = {"emp": "emp_17", "emp_fix": "emp_fix_17"}, inplace=True)
Mercer_ThreeDigitNAICS_Totals
```

# Convert index dtype to str

```python
Mercer_ThreeDigitNAICS_Totals.index = Mercer_ThreeDigitNAICS_Totals.index.astype(str)
Mercer_ThreeDigitNAICS_Totals.index.dtype
```

# Import NAICS index

```python
naics_index = pd.read_csv(r"G:\Shared drives\Community & Economic Development\Economic Clusters\CBP_ThreeDigitNAICS_Index.csv", index_col='NAICS')
naics_index.index= naics_index.index.astype(str)
naics_index.index.dtype
```

# Join dfs

```python
Mercer_NAICStotals = pd.merge(naics_index, Mercer_ThreeDigitNAICS_Totals, how="left", left_index=True, right_index=True)
Mercer_NAICStotals
```

```python
Mercer_NAICStotals.to_csv(r"G:\My Drive\BCarney\DataAnalysis\CBP_2017\ThreeDigit_NAICS\DVRPC_IndividualCounties\CBP2017_DVRPC_IndividualCounties_Mercer_NAICStotals.csv")
```
