```python
import pandas as pd
import numpy as np
```

# Import 2017 Data

```python
raw_2017 = pd.read_csv(r"G:\My Drive\BCarney\DataAnalysis\CBP_2017\NAICS_AllDigits\County_level\rawdata\cbp17co.txt")
raw_2017
```

# Select Three-Digit NAICS

```python
threedigit_NAICS = raw_2017[raw_2017['naics'].str.contains('///')]
threedigit_NAICS
```

# Create new three-digit NAICS column

```python
threedigit_NAICS['naics_threedigit'] = threedigit_NAICS['naics'].str.slice(start=0, stop=3)
threedigit_NAICS
```

# Check for empflag w/ three-digit NAICS

```python
threedigit_NAICS_empflagcheck = threedigit_NAICS[threedigit_NAICS.empflag.notnull()]
threedigit_NAICS_empflagcheck
```

# Select Gloucester County

```python
Gloucester_NAICS = threedigit_NAICS.loc[(threedigit_NAICS['fipstate'] == 34) & (threedigit_NAICS['fipscty'] == 15)]
Gloucester_NAICS
```

```python
Gloucester_ThreeDigitNAICS = Gloucester_NAICS[['naics_threedigit', 'empflag', 'emp']]
Gloucester_ThreeDigitNAICS
```

# Check for empflag

```python
empflag_threedigitcheck = Gloucester_ThreeDigitNAICS.loc[Gloucester_ThreeDigitNAICS['empflag'].notnull()]
empflag_threedigitcheck
```

# Create midpoint values for empflag rows

```python
Gloucester_midpoint = Gloucester_ThreeDigitNAICS.assign(emp_fix = Gloucester_ThreeDigitNAICS['emp'])
Gloucester_midpoint
```

```python
Gloucester_midpoint['emp_fix'].loc[(Gloucester_midpoint['empflag'] == 'A')] = 10
Gloucester_midpoint['emp_fix'].loc[(Gloucester_midpoint['empflag'] == 'B')] = 60
Gloucester_midpoint['emp_fix'].loc[(Gloucester_midpoint['empflag'] == 'C')] = 175
Gloucester_midpoint['emp_fix'].loc[(Gloucester_midpoint['empflag'] == 'E')] = 375
Gloucester_midpoint['emp_fix'].loc[(Gloucester_midpoint['empflag'] == 'F')] = 750
Gloucester_midpoint['emp_fix'].loc[(Gloucester_midpoint['empflag'] == 'G')] = 1750
Gloucester_midpoint['emp_fix'].loc[(Gloucester_midpoint['empflag'] == 'H')] = 3750
Gloucester_midpoint['emp_fix'].loc[(Gloucester_midpoint['empflag'] == 'I')] = 7500
Gloucester_midpoint['emp_fix'].loc[(Gloucester_midpoint['empflag'] == 'J')] = 17500
Gloucester_midpoint['emp_fix'].loc[(Gloucester_midpoint['empflag'] == 'K')] = 37500
Gloucester_midpoint['emp_fix'].loc[(Gloucester_midpoint['empflag'] == 'L')] = 75000
Gloucester_midpoint['emp_fix'].loc[(Gloucester_midpoint['empflag'] == 'M')] = 100000
Gloucester_midpoint
```

# View rows where emp_fix does not equal emp

```python
check = Gloucester_midpoint.loc[Gloucester_midpoint['emp'] != Gloucester_midpoint['emp_fix']]
check
```

# Group by 3-digit NAICS, sum employment

```python
Gloucester_ThreeDigitNAICS_Totals = Gloucester_midpoint.groupby('naics_threedigit').sum()
Gloucester_ThreeDigitNAICS_Totals
```

```python
Gloucester_ThreeDigitNAICS_Totals.rename(columns = {"emp": "emp_17", "emp_fix": "emp_fix_17"}, inplace=True)
Gloucester_ThreeDigitNAICS_Totals
```

# Change index dtype to str

```python
Gloucester_ThreeDigitNAICS_Totals.index = Gloucester_ThreeDigitNAICS_Totals.index.astype(str)
Gloucester_ThreeDigitNAICS_Totals.index.dtype
```

# Import NAICS index

```python
naics_index = pd.read_csv(r"G:\Shared drives\Community & Economic Development\Economic Clusters\CBP_ThreeDigitNAICS_Index.csv", index_col='NAICS')
naics_index
```

# Change index dtype to str

```python
naics_index.index = naics_index.index.astype(str)
naics_index.index.dtype
```

# Join two dfs

```python
Gloucester_NAICStotals = pd.merge(naics_index, Gloucester_ThreeDigitNAICS_Totals, how="left", left_index=True, right_index=True)
Gloucester_NAICStotals
```

```python
Gloucester_NAICStotals.to_csv(r"G:\My Drive\BCarney\DataAnalysis\CBP_2017\ThreeDigit_NAICS\DVRPC_IndividualCounties\CBP2017_DVRPC_IndividualCounties_Gloucester_NAICStotals.csv")
```
