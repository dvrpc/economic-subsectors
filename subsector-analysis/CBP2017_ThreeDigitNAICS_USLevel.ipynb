```python
import pandas as pd
import numpy as np
```

# Import 2017 US-Level Data

```python
raw_2017 = pd.read_csv(r"G:\My Drive\BCarney\DataAnalysis\CBP_2017\NAICS_AllDigits\US_level\rawdata\cbp17us.txt")
raw_2017
```

# Select Three-Digit NAICS

```python
threedigit_NAICS = raw_2017[raw_2017['naics'].str.contains('///')]
threedigit_NAICS
```

# Create new three-digit NAICS column

```python
threedigit_NAICS['naics_threedigit'] = threedigit_NAICS['naics'].str.slice(start=0, stop=3)
threedigit_NAICS
```

# Select All Employment rows (lfo = -)

```python
threedigit_NAICS = threedigit_NAICS.loc[threedigit_NAICS['lfo'] == '-']
threedigit_NAICS
```

# Check for empflag w/ three-digit NAICS

```python
threedigit_NAICS_empflagcheck = threedigit_NAICS[threedigit_NAICS.empflag.notnull()]
threedigit_NAICS_empflagcheck
```

```python
CBP2017_ThreeDigitNAICS = threedigit_NAICS[['naics_threedigit', 'empflag', 'emp']]
CBP2017_ThreeDigitNAICS
```

```python
empflag_threedigitcheck = CBP2017_ThreeDigitNAICS.loc[CBP2017_ThreeDigitNAICS['empflag'].notnull()]
empflag_threedigitcheck
```

```python
ThreeDigitNAICS_Totals = CBP2017_ThreeDigitNAICS.groupby('naics_threedigit').sum()
ThreeDigitNAICS_Totals
```

```python
ThreeDigitNAICS_Totals.to_csv(r"G:\My Drive\BCarney\DataAnalysis\CBP_2017\ThreeDigit_NAICS\AllUS_USLevel\CBP2017_ThreeDigitNAICSTotal_AllUS_USLevel.csv")
```
