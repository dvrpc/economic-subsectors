```python
import pandas as pd
import numpy as np
```

# Import 2017 US-Level Data

```python
raw_2017 = pd.read_csv(r"G:\My Drive\BCarney\DataAnalysis\CBP_2017\NAICS_AllDigits\County_level\rawdata\cbp17co.txt")
raw_2017
```

# Select Three-Digit NAICS

```python
threedigit_NAICS = raw_2017[raw_2017['naics'].str.contains('///')]
threedigit_NAICS
```

# Create new three-digit NAICS column

```python
threedigit_NAICS['naics_threedigit'] = threedigit_NAICS['naics'].str.slice(start=0, stop=3)
threedigit_NAICS
```

# Check for empflag w/ three-digit NAICS

```python
threedigit_NAICS_empflagcheck = threedigit_NAICS[threedigit_NAICS.empflag.notnull()]
threedigit_NAICS_empflagcheck
```

# Create midpoint values for empflag rows

```python
threedigit_NAICS = threedigit_NAICS.assign(emp_fix = threedigit_NAICS['emp'])
threedigit_NAICS
```

```python
threedigit_NAICS['emp_fix'].loc[(threedigit_NAICS['empflag'] == 'A')] = 10
threedigit_NAICS['emp_fix'].loc[(threedigit_NAICS['empflag'] == 'B')] = 60
threedigit_NAICS['emp_fix'].loc[(threedigit_NAICS['empflag'] == 'C')] = 175
threedigit_NAICS['emp_fix'].loc[(threedigit_NAICS['empflag'] == 'E')] = 375
threedigit_NAICS['emp_fix'].loc[(threedigit_NAICS['empflag'] == 'F')] = 750
threedigit_NAICS['emp_fix'].loc[(threedigit_NAICS['empflag'] == 'G')] = 1750
threedigit_NAICS['emp_fix'].loc[(threedigit_NAICS['empflag'] == 'H')] = 3750
threedigit_NAICS['emp_fix'].loc[(threedigit_NAICS['empflag'] == 'I')] = 7500
threedigit_NAICS['emp_fix'].loc[(threedigit_NAICS['empflag'] == 'J')] = 17500
threedigit_NAICS['emp_fix'].loc[(threedigit_NAICS['empflag'] == 'K')] = 37500
threedigit_NAICS['emp_fix'].loc[(threedigit_NAICS['empflag'] == 'L')] = 75000
threedigit_NAICS['emp_fix'].loc[(threedigit_NAICS['empflag'] == 'M')] = 100000
threedigit_NAICS
```

```python
check = threedigit_NAICS.loc[threedigit_NAICS['emp'] != threedigit_NAICS['emp_fix']]
check
```

```python
CBP2017_ThreeDigitNAICS = threedigit_NAICS[['naics_threedigit', 'emp', 'emp_fix']]
CBP2017_ThreeDigitNAICS
```

```python
ThreeDigitNAICS_Totals = CBP2017_ThreeDigitNAICS.groupby('naics_threedigit').sum()
ThreeDigitNAICS_Totals.rename(columns = {'emp': 'emp_17', 'emp_fix': 'emp_fix_17'}, inplace=True)
ThreeDigitNAICS_Totals
```

# Convert Index to String dtype

```python
ThreeDigitNAICS_Totals.index = ThreeDigitNAICS_Totals.index.astype(str, copy = False)
```

# Import NAICS Index

```python
naics_index = pd.read_csv(r"G:\Shared drives\Community & Economic Development\Economic Clusters\CBP_ThreeDigitNAICS_Index.csv", index_col='NAICS')
```

# Convert Index to String dtype

```python
naics_index.index = naics_index.index.astype(str, copy = False)
```

# Join dfs

```python
AllUS_NAICStotals = pd.merge(naics_index, ThreeDigitNAICS_Totals, how="left", left_index=True, right_index=True)
AllUS_NAICStotals
```
