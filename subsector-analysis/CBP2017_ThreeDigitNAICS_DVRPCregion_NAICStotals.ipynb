```python
import pandas as pd
import numpy as np
```

# Import 2017 Data

```python
raw_2017 = pd.read_csv(r"G:\My Drive\BCarney\DataAnalysis\CBP_2017\NAICS_AllDigits\County_level\rawdata\cbp17co.txt")
raw_2017
```

# Select Three-Digit NAICS

```python
threedigit_NAICS = raw_2017[raw_2017['naics'].str.contains('///')]
threedigit_NAICS
```

# Create new three-digit NAICS column

```python
threedigit_NAICS['naics_threedigit'] = threedigit_NAICS['naics'].str.slice(start=0, stop=3)
threedigit_NAICS
```

```python
threedigit_NAICS_empflagcheck = threedigit_NAICS[threedigit_NAICS.empflag.notnull()]
threedigit_NAICS_empflagcheck
```

# Select PA and NJ

```python
NJandPA = threedigit_NAICS[(threedigit_NAICS['fipstate'] == 34) | (threedigit_NAICS['fipstate'] == 42)]
NJandPA
```

# Select DVRPC Region

```python
region_NAICS = NJandPA.loc[((NJandPA['fipstate'] == 34) & ((NJandPA['fipscty'] == 5) | (NJandPA['fipscty'] == 7) | (NJandPA['fipscty'] == 15) | (NJandPA['fipscty'] == 21))) |  ((NJandPA['fipstate'] == 42) & ((NJandPA['fipscty'] == 17) | (NJandPA['fipscty'] == 29) | (NJandPA['fipscty'] == 45) | (NJandPA['fipscty'] == 91) | (NJandPA['fipscty'] == 101)))]
region_NAICS
```

```python
CBP2017_ThreeDigitNAICS = region_NAICS[['naics_threedigit', 'empflag', 'emp']]
CBP2017_ThreeDigitNAICS
```

```python
empflag_threedigitcheck = CBP2017_ThreeDigitNAICS.loc[CBP2017_ThreeDigitNAICS['empflag'].notnull()]
empflag_threedigitcheck
```

# Create midpoint values for empflag rows

```python
ninecountyregion = CBP2017_ThreeDigitNAICS.assign(emp_fix = CBP2017_ThreeDigitNAICS['emp'])
ninecountyregion
```

```python
ninecountyregion['emp_fix'].loc[(ninecountyregion['empflag'] == 'A')] = 10
ninecountyregion['emp_fix'].loc[(ninecountyregion['empflag'] == 'B')] = 60
ninecountyregion['emp_fix'].loc[(ninecountyregion['empflag'] == 'C')] = 175
ninecountyregion['emp_fix'].loc[(ninecountyregion['empflag'] == 'E')] = 375
ninecountyregion['emp_fix'].loc[(ninecountyregion['empflag'] == 'F')] = 750
ninecountyregion['emp_fix'].loc[(ninecountyregion['empflag'] == 'G')] = 1750
ninecountyregion['emp_fix'].loc[(ninecountyregion['empflag'] == 'H')] = 3750
ninecountyregion['emp_fix'].loc[(ninecountyregion['empflag'] == 'I')] = 7500
ninecountyregion['emp_fix'].loc[(ninecountyregion['empflag'] == 'J')] = 17500
ninecountyregion['emp_fix'].loc[(ninecountyregion['empflag'] == 'K')] = 37500
ninecountyregion['emp_fix'].loc[(ninecountyregion['empflag'] == 'L')] = 75000
ninecountyregion['emp_fix'].loc[(ninecountyregion['empflag'] == 'M')] = 100000
ninecountyregion
```

```python
check = ninecountyregion.loc[ninecountyregion['emp'] != ninecountyregion['emp_fix']]
check
```

```python
ThreeDigitNAICS_Totals = ninecountyregion.groupby('naics_threedigit').sum()
ThreeDigitNAICS_Totals.rename(columns = {"naics_threedigit": "naics_2017", "emp": "emp_2017", "emp_fix": "emp_fix_2017"}, inplace=True)
ThreeDigitNAICS_Totals
```

# Convert index dtype to string

```python
ThreeDigitNAICS_Totals.index = ThreeDigitNAICS_Totals.index.astype(str, copy = False)
```

# Import NAICS Index

```python
naics_index = pd.read_csv(r"G:\Shared drives\Community & Economic Development\Economic Clusters\CBP_ThreeDigitNAICS_Index.csv", index_col='NAICS')
naics_index
```

# Convert Index dtype

```python
naics_index.index = naics_index.index.astype(str, copy = False)
```

# Join dfs

```python
DVRPC_NAICStotals = pd.merge(naics_index, ThreeDigitNAICS_Totals, how="left", left_index=True, right_index=True)
DVRPC_NAICStotals
```

```python
DVRPC_NAICStotals.to_csv(r"G:\My Drive\BCarney\DataAnalysis\CBP_2017\ThreeDigit_NAICS\DVRPC_region\CBP2017_ThreeDigitNAICS_DVRPCregion_NAICStotals.csv")
```
