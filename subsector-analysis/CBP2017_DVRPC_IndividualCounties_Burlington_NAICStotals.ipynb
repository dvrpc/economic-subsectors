```python
import pandas as pd
import numpy as np
```

# Import 2017 Data

```python
raw_2017 = pd.read_csv(r"G:\My Drive\BCarney\DataAnalysis\CBP_2017\NAICS_AllDigits\County_level\rawdata\cbp17co.txt")
raw_2017
```

# Select Three-Digit NAICS

```python
threedigit_NAICS = raw_2017[raw_2017['naics'].str.contains('///')]
threedigit_NAICS
```

# Create new three-digit NAICS column

```python
threedigit_NAICS['naics_threedigit'] = threedigit_NAICS['naics'].str.slice(start=0, stop=3)
threedigit_NAICS
```

# Check for empflag w/ three-digit NAICS

```python
threedigit_NAICS_empflagcheck = threedigit_NAICS[threedigit_NAICS.empflag.notnull()]
threedigit_NAICS_empflagcheck
```

# Select Burlington County

```python
Burlington_NAICS = threedigit_NAICS.loc[(threedigit_NAICS['fipstate'] == 34) & (threedigit_NAICS['fipscty'] == 5)]
Burlington_NAICS
```

```python
Burlington_ThreeDigitNAICS = Burlington_NAICS[['naics_threedigit', 'empflag', 'emp']]
Burlington_ThreeDigitNAICS
```

# Check for empflag

```python
empflag_threedigitcheck = Burlington_ThreeDigitNAICS.loc[Burlington_ThreeDigitNAICS['empflag'].notnull()]
empflag_threedigitcheck
```

# Create midpoint values for empflag rows

```python
Burlington_midpoint = Burlington_ThreeDigitNAICS.assign(emp_fix = Burlington_ThreeDigitNAICS['emp'])
Burlington_midpoint
```

```python
Burlington_midpoint['emp_fix'].loc[(Burlington_midpoint['empflag'] == 'A')] = 10
Burlington_midpoint['emp_fix'].loc[(Burlington_midpoint['empflag'] == 'B')] = 60
Burlington_midpoint['emp_fix'].loc[(Burlington_midpoint['empflag'] == 'C')] = 175
Burlington_midpoint['emp_fix'].loc[(Burlington_midpoint['empflag'] == 'E')] = 375
Burlington_midpoint['emp_fix'].loc[(Burlington_midpoint['empflag'] == 'F')] = 750
Burlington_midpoint['emp_fix'].loc[(Burlington_midpoint['empflag'] == 'G')] = 1750
Burlington_midpoint['emp_fix'].loc[(Burlington_midpoint['empflag'] == 'H')] = 3750
Burlington_midpoint['emp_fix'].loc[(Burlington_midpoint['empflag'] == 'I')] = 7500
Burlington_midpoint['emp_fix'].loc[(Burlington_midpoint['empflag'] == 'J')] = 17500
Burlington_midpoint['emp_fix'].loc[(Burlington_midpoint['empflag'] == 'K')] = 37500
Burlington_midpoint['emp_fix'].loc[(Burlington_midpoint['empflag'] == 'L')] = 75000
Burlington_midpoint['emp_fix'].loc[(Burlington_midpoint['empflag'] == 'M')] = 100000
Burlington_midpoint
```

# View rows where emp_fix does not equal emp

```python
check = Burlington_midpoint.loc[Burlington_midpoint['emp'] != Burlington_midpoint['emp_fix']]
check
```

# Group by 3-digit NAICS and sum employment

```python
Burlington_ThreeDigitNAICS_Totals = Burlington_midpoint.groupby('naics_threedigit').sum()
Burlington_ThreeDigitNAICS_Totals
```

```python
Burlington_ThreeDigitNAICS_Totals.rename(columns = {"emp": "emp_17", "emp_fix": "emp_fix_17"}, inplace=True)
Burlington_ThreeDigitNAICS_Totals
```

# Convert index dtype to str

```python
Burlington_ThreeDigitNAICS_Totals.index = Burlington_ThreeDigitNAICS_Totals.index.astype(str)
Burlington_ThreeDigitNAICS_Totals.index.dtype
```

# Import NAICS index

```python
naics_index = pd.read_csv(r"G:\Shared drives\Community & Economic Development\Economic Clusters\CBP_ThreeDigitNAICS_Index.csv", index_col='NAICS')
naics_index
```

# Convert index dtype to str

```python
naics_index.index = naics_index.index.astype(str)
naics_index.index.dtype
```

# Join two dfs

```python
Burlington_NAICStotals = pd.merge(naics_index, Burlington_ThreeDigitNAICS_Totals, how="left", left_index=True, right_index=True)
Burlington_NAICStotals
```

```python
Burlington_NAICStotals.to_csv(r"G:\My Drive\BCarney\DataAnalysis\CBP_2017\ThreeDigit_NAICS\DVRPC_IndividualCounties\CBP2017_DVRPC_IndividualCounties_Burlington_NAICStotals.csv")
```
